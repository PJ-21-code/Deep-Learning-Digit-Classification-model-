Resizing the image

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resize= img.resize((200,200))
```

```python
img_resize.save('dog_image_resized.jpg')
```

```python
# displaying image from numpy array
img_resize= mpimg.imread('/content/dog_image_resized.jpg')
img_plot_resized= plt.imshow(img_resize)
plt.show()
```

converting rgb image to grayscale image using openCV

```python
import cv2
```

```python
img= cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image= cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
plt.imshow(grayscale_image)
plt.show()
```

cv2.imshow() will show image

```python
from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
#saving grayscale image
cv2.imwrite('dog_grayscale_image.jpg',grayscale_image)
```
